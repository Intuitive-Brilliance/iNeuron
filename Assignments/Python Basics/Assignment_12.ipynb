#### 4. If a PdfFileReader object’s PDF is encrypted with the password swordfish, what must you do before you can obtain Page objects from it?
**Ans:** 
<br>First read the Pdf using the PdfFileReader Class. 
<br>PdfFileReader class provides a attribute called **`isEncrypted`** to check whether a pdf is encrypted or not. 
<br>The method returns true if a pdf is encrypted.
<br>If pdf is encrypted use the **`decrypt()`** method provided by PdfFileReader class first then try to read the contents/pages of the pdf, else PyPDF2 will raise the following error **`PyPDF2.utils.PdfReadError: file has not been decrypted`**

```python
#Example Code:
from PyPDF2 import PdfFileReader
pdf_reader = PdfFileReader("01_Lange_Michelle_mapld09_pres_1.pdf")
if pdf_reader.isEncrypted: # to check whether the pdf is encrypted or not
    pdf_reader.decrypt("swordfish")
for page in pdf_reader.pages:
    print(page.extractText()) # to print the text data of a page from pdf
```

#### 6. What is the difference between a Run object and a Paragraph object?
**Ans:** 
<br>A Document object contains a list of **`Paragraph`** objects for the paragraphs in the document. 
<br>A new paragraph begins whenever the user presses `ENTER` or `RETURN` while typing in a Word document.
<br>Each of these Paragraph objects contains a list of one or more **`Run`** objects.
<br>A Run object is a contiguous run of text with the same style. 
<br>A new Run object is needed whenever the text style changes.
<br>This is because a text in a Word document is more than just a string. It has font, size, color, and other styling information associated with it. A style in Word is a collection of these attributes. 

#### 7. How do you obtain a list of Paragraph objects for a Document object that’s stored in a variable named doc?

```python
# Example
from docx import Document
doc = Document("report.docx") # Path of the file
print(doc.paragraphs) # Prints the list of Paragraph objects for a Document
for paragraph in doc.paragraphs:
    print(paragraph.text) # Prints the text in the paragraph
```
